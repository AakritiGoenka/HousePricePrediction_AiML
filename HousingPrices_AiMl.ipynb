```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns #better visualisation over matplotlib, advanced visualisation
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
df=pd.read_csv('boston.csv')
#ds=sklearn.datasets.load_boston() this was valid before because boston dataset was preincluded with scikit learn 

```

```python
df.head()
```

```python
df['price']=df.MEDV
```

```python
df.head()
```

```python
df.drop(columns=['MEDV'],inplace=True)
```

```python
df.head()
```

```python
df.shape

```

```python
#check for missing values
df.isnull().sum()
```

```python
df.describe()
```

Understanding the correlations between various features
Positive Correlation= one increases other increases
Negative Corrrelation=one increases other decreases


```python
correlation=df.corr()
```

```python
#to find correlation we construct a heat map
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
#cbar=colorbar in the side 
```

```python
X=df.drop(columns=['price'],axis=1)
```

```python
Y=df['price']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
#loading model
model=XGBRegressor()
```

```python
#training model
model.fit(X_train,Y_train)
```

```python
#Prediction
pred=model.predict(X_train)
print(pred)
```

```python
#Evaluating errors
#for classification models:accuracy score
#for regression models: Rsq and mean abs error etc

#R2 error
score_1=metrics.r2_score(Y_train,pred)

#meean abs error
score_2=metrics.mean_absolute_error(Y_train,pred)
```

```python
print("R square error")
print(score_1)
#should be close to 1 not 10 so basically shld be small
```

```python
print("Mean absolute error")
print(score_2)
```

```python
pred_test=model.predict(X_test)
```

```python
#R2 error
score_t1=metrics.r2_score(Y_test,pred_test)

#meean abs error
score_t2=metrics.mean_absolute_error(Y_test,pred_test)
print("R square error")
print(score_t1)
print("Mean absolute error")
print(score_t2)
```

Visualising actual and predicted prices
